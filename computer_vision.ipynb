````python
import cv2
import numpy as np
import matplotlib.pyplot as plt
````

````python
a = cv2.imread('froggy.jpg')
print(a.shape)
````

````python
plt.imshow(a)
````

````python
b = cv2.cvtColor(a, cv2.COLOR_BGR2GRAY)
plt.imshow(b, cmap='gray')
````

````python
c = cv2.cvtColor(a, cv2.COLOR_BGR2RGB)
plt.imshow(c)
````

````python
# resize by resolution
print(a.shape)
cr = cv2.resize(c, (1600, 1500))
plt.imshow(cr)
````

````python
# resize by scale
cscaled = cv2.resize(c, None, fx=0.5, fy=0.5) # image reduced by 50%
plt.imshow(cscaled)
````

````python
# apply blur
cblur = cv2.GaussianBlur(c, (11,51), 0)
plt.imshow(cblur)
# save
cblur = cv2.cvtColor(cblur, cv2.COLOR_RGB2BGR)
cv2.imwrite('blurred_froggy.jpg', cblur)
````

````python
# apply edge detection
cgray = cv2.cvtColor(c, cv2.COLOR_RGB2GRAY)
cgray = cv2.GaussianBlur(cgray, (5,5), 0)
cgray = cv2.Canny(cgray, 10, 150)
plt.imshow(cgray, cmap='gray')
````

````python
# invert image
cinvert = cv2.bitwise_not(cgray)
plt.imshow(cinvert, cmap='gray')
````
